```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Dataset
# Replace this with:
# anime_data_extracted = pd.read_csv("anime.csv")

data = {
    "Name": [
        "Naruto",
        "Attack on Titan",
        "Your Name",
        "One Piece",
        "Death Note",
        "Demon Slayer"
    ],
    "Type": [
        "TV",
        "TV",
        "Movie",
        "TV",
        "TV",
        "Movie"
    ],
    "Score": [
        7.9,
        9.1,
        8.9,
        8.7,
        9.0,
        8.8
    ]
}

anime_data_extracted = pd.DataFrame(data)

# Function
def homework(anime_data_extracted):

    # Group by anime type and calculate average score
    result = (
        anime_data_extracted
        .groupby('Type')['Score']
        .mean()
        .sort_values(ascending=False)
    )

    return result

# Run function
result = homework(anime_data_extracted)

# Print results
print("Average Anime Scores by Type:\n")
print(result)

# Visualization
plt.figure(figsize=(7,5))

sns.barplot(
    x=result.values,
    y=result.index
)

plt.title("Average Anime Score by Type")
plt.xlabel("Average Score")
plt.ylabel("Anime Type")

plt.show()
```
